```python
import numpy as np
import pandas as pd
# Beautiful Soup is a Python library for pulling data out of HTML and XML files.
from bs4 import BeautifulSoup
# The requests module allows you to send HTTP requests using Python.
import requests
# The pprint module provides a capability to “pretty-print” arbitrary Python
# data structures in a form which can be used as input to the interpreter.
from pprint import pprint
# The OS module in Python provides functions for creating and removing a directory (folder).
import os
# This module offers classes representing filesystem paths with semantics appropriate for different operating systems.
from pathlib import Path
# The urllib.request module defines functions and classes which help in opening URLs 
# (mostly HTTP) in a complex world — basic and digest authentication, redirections, cookies and more.
import urllib.request
# This module provides various time-related functions.
import time


dirpath = ''

baseurl = "https://readcomicsonline.ru/comic"
# topicurl = "autobiographix-2021"
# topicurl = "always-never-2021"
# topicurl = "1984-the-graphic-novel-2021"
# topicurl = "arya-2017"
# topicurl = "atomic-robo-and-the-dawn-of-a-new-era-2019"
# topicurl = "a-tale-of-two-arthurs-2020"
# topicurl = "batman-noir-hush-2015"
# topicurl = "castaways-2022"
# topicurl = "david-bowie-in-comics-2022"
# topicurl = "dixie-road-2017"
# topicurl = "dead-eyes-2019"
# topicurl = "destiny-ny-summer-special-2020"
# topicurl = "cyberpunk-2077-blackout-2022"
# topicurl = "cyberpunk-2077-your-voice-2021"
# topicurl = "days-of-hate-2018"
# topicurl = "a-righteous-thirst-for-vengeance-2021"
# topicurl = "a-lapse-in-judgment-2020"
# topicurl = "dark-blood-2021"
# topicurl = "dune-the-graphic-novel-2020"
# topicurl = "ether-copper-golems-2018"

topicurl = "clandestino-2015"

fullurl = "/".join([baseurl, topicurl])
fullurl
```

```python
topic_dir = os.path.join(dirpath, topicurl)
Path(topic_dir).mkdir(parents=True, exist_ok=True)
page = requests.get(fullurl)

if page.status_code==200:
    
    soup = BeautifulSoup(page.content, 'lxml')
    issues = [int(el.a.text.split("#")[-1]) for el in soup.findAll('h5', attrs={'class' : 'chapter-title-rtl'})]
    issues.sort()
    for issue in issues:
        issue_dir = os.path.join(topic_dir, str(issue))
        Path(issue_dir).mkdir(parents=True, exist_ok=True)
        issueurl = "/".join([fullurl, str(issue)])
        print(issueurl)
        k=1
        while True:
            print(issue, k)
            pageurl = "/".join([issueurl, str(k)])
            subpage = requests.get(pageurl)
            subsoup = BeautifulSoup(subpage.content, 'lxml')
            print(subpage.status_code)
            if subpage.status_code!=200: break
            el = subsoup.findAll('img', attrs={'class' : 'img-responsive scan-page'})
            srcurl = el[0]['src']
            img_path = os.path.join(issue_dir, str(k))+".jpg"
            opener = urllib.request.build_opener()
            opener.addheaders = [('User-Agent', 'MyApp/1.0')]
            urllib.request.install_opener(opener)
            urllib.request.urlretrieve(srcurl.strip(), img_path)
            k = k+1
            time.sleep(3)
else:
    print(page.status_code)
```
